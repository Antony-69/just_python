Використовуючи цикл for, виведіть числа від 1 до 10, але припиніть виконання при досягненні числа 7 (оператор break).

```python
for i in range (1, 10):
    if i>7:
        break
    print(i)
```

Створіть програму, яка виводить числа від 1 до 10, пропускаючи парні числа (оператор continue).

```python
for i in range(1, 10):
    if i % 2 == 0:
        continue
    print(i)
```

Напишіть програму, яка запитує у користувача пароль. Якщо пароль правильний – програма завершує цикл (break), якщо неправильний – знову просить ввести.

```python
c_password = "marquis"

while True:
    password = input("Enter your password: ")
    if password == c_password:
        print("Access granted")
        break
    else:
        print("Access denied. Try again.")
```

Реалізуйте програму для обчислення суми чисел від 1 до 100, але з використанням continue пропустіть всі кратні 5 числа.

```python
for i in range(1, 100):
    if i % 5== 0:
        continue
    i += i
print(i)
```

Створіть програму, яка виводить всі непарні числа від 1 до 20, використовуючи цикл while, оператори continue та break.

```python
i = 0
while True:
    i += 1
    if i % 2 == 0:
        continue
    if i > 20:
        break
    print(i)
```

Напишіть програму, яка знаходить перше число більше 50, яке ділиться і на 3, і на 7 (використати while і break).

```python
i = 50
while True:
    i += 1
    if i % 3 == 0 and i % 7 == 0:
        print(i)
        break
    
```
