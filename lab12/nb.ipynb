<ім’я>!». Викличте її кілька разів з різними аргументами.

Створіть функцію square(x), яка повертає квадрат числа.

```python
def square(x):
    return x*x
print(square(5))
print(square(9))
```

Створіть функцію sum_numbers(a, b), яка повертає суму двох чисел.

```python
def sum_numbers(a, b):
    return a + b

print(sum_numbers(3, 4))
```

Напишіть функцію max_of_two(a, b), яка повертає більше з двох чисел.

```python
def max_of_two(a, b):
    if a > b:
        return a
    else:
        return b
print(max_of_two(10, 20))
```

Реалізуйте функцію convert(celsius), яка переводить температуру з градусів
Цельсія у Фаренгейти за формулою:

```python
def convert(celsius):
    return (celsius * 9/5) + 32
celsius = int(input("Enter temperature in Celsius: "))
fahrenheit = convert(celsius)
print(f"{celsius}°C is {fahrenheit}°F")
```

Напишіть функцію circle(r), яка обчислює і повертає площу та довжину кола
для заданого радіуса.

```python
def circle(r):
    pi = 3.14159
    print(f"Area of circle with radius {r} is {pi * r * r}, circumference is {2 * pi * r}")
r = float(input("Enter radius: "))
circle(r)
```

Створіть функцію is_even(n), яка повертає True, якщо число парне, і False,
якщо непарне.

```python
def is_even(n):
    return n % 2 == 0
n = int(input("Enter a number: "))
print(is_even(n))
```
