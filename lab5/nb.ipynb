Напишіть програму, яка конвертує кілометри у метри.

```python
km = float(input("Enter the distance (km)"))
m = km * 1000
print(f"{km} km is {m} meters")
```

Створіть програму-конвертер для переведення градусів Цельсія у Фаренгейти

```python
Celsius = float(input("Enter temperature in Celsius: "))
Fahrenheit = (Celsius * 9/5) + 32
print(f"{Celsius}°C is {Fahrenheit}°F")
```

Реалізуйте програму для перетворення годин у хвилини та секунди.

```python
h = int(input("Enter hours: "))
m = h * 60
s = h * 3600
print(f"{h} hours is {m} minutes or {s} seconds")
```

Створіть конвертер валют (наприклад, гривні у долари) з використанням
змінної для курсу обміну.

```python
#uah to mnt (tugrik)
uah = float(input("Enter the amount in UAH"))
mnt = round(uah * 84.1607, 2)
print(f"{uah} UAH is {mnt} tugriks")
```

Напишіть програму, яка зчитує від користувача значення в сантиметрах і
переводить його у метри та кілометри.

```python
cm = float(input("Enter the length in centimeters: "))
m = cm / 100
km = cm / 100000
print(f"{cm} cm is {m} meters or {km} kilometers")
```
