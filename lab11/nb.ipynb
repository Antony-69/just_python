Напишіть програму, яка перевіряє, чи міститься введене користувачем слово
у списку слів.

```python
list = []
for i in range (3):
    value = str(input("Enter a word: "))
    list.append(value)
print (list)

print ("Marquis" in list)


#or the other way around

list = ["Mercury", "Grand", "Marsquis"]
word = str(input("Enter a word: "))
print (word in list)
```
