Створіть список із чисел і відсортуйте його за зростанням.

```python
list = [1, 8, 9, 4, 5]
list.sort()
print (list)
```

Створіть список із чисел і відсортуйте його за спаданням.

```python
list = [1, 8, 9, 4, 5]
list.sort(reverse=True)
print (list)
```

Створіть список із кількох слів і відсортуйте його в алфавітному порядку.

```python
fruits = ["apple", "cherry", "banana"]
fruits.sort()
print(fruits)
```

Створіть список із рядків і відсортуйте його за довжиною елементів.

```python
fruits = ["apple", "strawberry", "banana"]
fruits.sort(key=len)
print(fruits)
```

Напишіть програму, яка зчитує з клавіатури 5 чисел, додає їх у список, а
потім виводить список у відсортованому порядку.

```python
list = []
for i in range (5):
    value = int(input("Enter a number: "))
    list.append(value)
print (list)

list.sort()
print(list)
```
