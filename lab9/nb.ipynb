Створіть список, який містить кілька однакових елементів. Використайте методи count() і index(), щоб визначити, скільки разів зустрічається елемент і на якій позиції він з’являється вперше.

```python
list = [7, 9, 7, 56, 78, 2, 485, 7, 2, 7, 1, 9, 3, 9, 4, 8]
for i in set(list):
    count = list.count(i)
    first_index = list.index(i)
    print(f"Value {i} occurs {count} times, first at index {first_index}")
    
```
