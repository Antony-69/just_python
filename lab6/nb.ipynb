Напишіть програму, яка перевіряє, чи введене число є додатним.

````python
n = float(input("Enter a number:") )
if n > 0 :
    print(f"{n} is a positive number")
else:
    print(f"{n} is not a positive number")
````

Створіть програму, яка визначає, чи введене число парне чи непарне.

````python
n = float(input("Enter a number:") )
if n%2 == 0 :
    print(f"{n} is an even number")
else:
    print(f"{n} is an odd number")
````

Реалізуйте програму, яка зчитує два числа та виводить більше з них.

````python
a = int(input("Enter a number:") )
b = int(input("Enter another number:") )
if a > b:
    print(f"{a} is greater than {b} \n {a}")
elif a < b:
    print(f"{a} is less than {b} \n {b}")
else:
    print(f"{a} is equal to {b}")
````
