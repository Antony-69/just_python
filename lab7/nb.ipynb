Напишіть програму, яка перевіряє, чи два введених числа рівні

```python
a = int(input("Enter a value: "))
b = int(input("And another one: "))

if a==b:
    print(f"{a} is equal to {b}")
else:
    print(f"{a} is not equal to {b}")
```

Створіть програму, яка перевіряє, чи введене число більше нуля

```python
a = float(input("Enter a number:") )
if a > 0 :
    print(f"{a} is a positive number")
else:
    print(f"{a} is not a positive number")
```

Напишіть програму, яка порівнює два рядки та виводить, чи вони однакові.

```python
str1 = input("Enter first string: ")
str2 = input("Enter second string: ")
if str1 == str2:
    print("The strings are equal")
else:
    print("The strings are not equal")
```

Реалізуйте програму, яка перевіряє, чи число входить у діапазон від 1 до 100 включно.

```python
a = float(input("Enter a number:") )
if a>=1 and a<=100:
    print(f"{a} is between 1 and 100")
```

Створіть програму, яка перевіряє логін користувача: якщо введене значення збігається зі змінною login = "admin", виведіть “Доступ дозволено”, інакше – “Доступ заборонено”.

```python
c_login = "admin"

login = input("Enter your login: ")
if login == c_login:
    print("Access granted")
else:
    print("Access denied")
```

Напишіть програму, яка приймає три числа і перевіряє, чи розташовані вони у порядку зростання.

```python
a = float(input("Enter a first number:") )
b = float(input("Enter a second number:") )
c = float(input("Enter a third number:") )

if a < b < c:
    print("The numbers are in ascending order")
else:
    print("The numbers are not in ascending order")
```
