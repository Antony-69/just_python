Напишіть програму, яка перевіряє, чи число, введене користувачем, є
парним.

````python
input ("Input your number: ")
number = int(input("Input your number: "))
print(f"You entered: {number}")
if number % 2 == 0:
    print("Your number is even")
else:
    print("Your number is odd")
````

Створіть програму, яка обчислює суму квадратів двох чисел,
використовуючи оператор присвоєння **=.

````python
list = [3, 4]
list[0] **= 2
list[1] **= 2
sum_squares = list[0] + list[1]
print(f'Squares: {list}, Sum of squares: {sum_squares}')
````
