Напишіть програму, яка додає нову пару «ключ: значення» у словник і
виводить оновлений словник.

```python
def add_to_dict(d, key, value):
    d[key] = value
    return d
my_dict = {"a": 1, "b": 2}
new_key = "c"
new_value = 3
updated_dict = add_to_dict(my_dict, new_key, new_value)
print(updated_dict)
```

Створіть словник із кількома словами англійською мовою як ключами та їх
перекладом українською. Реалізуйте «міні-словник»: користувач вводить
слово англійською, програма повертає переклад.

```python
dictionary = {
    "apple": "яблуко",
    "banana": "банан",
    "orange": "апельсин"
}
word = input("Enter an English word: ")
if word in dictionary:
    print(dictionary[word])
else:
    print("Word was not found in dictionary.")
```

Створіть словник товарів (назва → ціна). Напишіть програму, яка рахує
загальну вартість покупки, якщо користувач вводить кількість кожного
товару.

```python
products = {
    "apple": 12,
    "banana": 70,
    "orange": 80
}
total_cost = 0
for product, price in products.items():
    quantity = int(input(f"Enter quantity of {product}: "))
    total_cost += price * quantity
print(f"Total cost of purchase: {total_cost}")
```
